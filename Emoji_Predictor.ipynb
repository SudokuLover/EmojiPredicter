# STEP - 3 CONVERT SENTENCES INTO EMBEDDINGS

DOWNLOAD GLOVE6B50D.txt

```python
f = open('glove.6B.50d.txt',encoding = "utf-8")
```

```python
embedding_index = {}
count = 0
for line in f:
    values = line.split()
    word = values[0]
    coef = np.asarray(values[1:],dtype = 'float32')
    embedding_index[word] =coef 
"""print(word,coef)
   count += 1
    if count == 5:
        break """
```

```python
embedding_index['eat']
```

```python
emb_dim  = embedding_index['eat'].shape[0]
emb_dim
```

**STEP - 4 CONVERTING SENTENCE INTO VECTORS**

```python
def empbedding_output(x):
    
    maxLen = 10 
    embedding_out = np.zeros((x.shape[0],maxLen,emb_dim))
    for ix in range(x.shape[0]):
        x[ix] = x[ix].split()
        for ij in range(len(x[ix])):
            #go to every word in the current x[i] sentence
            try :
                embedding_out[ix][ij] = embedding_index[x[ix][ij].lower()]
            except:
                embedding_out[ix][ij] = np.zeros(emb_dim)
    return embedding_out
```

```python
embedding_matrix_train = empbedding_output(x_train)
embedding_matrix_test = empbedding_output(x_test)
```

```python
print(x_train[0])
print(len(x_train[0]))
```

```python
print(embedding_matrix_train.shape)
print(embedding_matrix_test.shape)
# got our embedding matrix
```

```python
# converting y_train into one_hot_encoding
from keras.utils import to_categorical 
```

```python
y_train = to_categorical(y_train , num_classes=5)
y_test = to_categorical(y_test , num_classes=5)
y_train.shape, y_test.shape
```

**STEP-5 DEFINE THE RNN/LSTM MODEL**

```python
# import models and layers
from keras.models import Sequential
from keras.layers import *
```

```python
# implementing Stack LSTM
model = Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences = True)) # 64 is a hidden unit
model.add(Dropout(0.5)) # regularization
model.add(LSTM(64,return_sequences = False))
model.add(Dropout(0.5))
model.add(Dense(5)) #output layer
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer= 'adam',metrics = ['accuracy'])
model.summary()
```

```python
# for early stop
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkPoint = ModelCheckpoint("best_model.h5", monitor= 'val_loss',verbose=True,save_best_only = True)
# it will work on val_loss matrix, save only best paramter and left intermidiate parameters
earlyStop = EarlyStopping(monitor = 'val_acc',patience=10)
# wait for changes in val_acc for 10 epoch else break.



hist = model.fit(embedding_matrix_train,y_train,epochs=100,batch_size=64,shuffle=True,validation_split=0.2)
# here we are oberfitting(aaccuravy = 1) as we have less data -> so to prevent overfitting need to do earlystop
```
